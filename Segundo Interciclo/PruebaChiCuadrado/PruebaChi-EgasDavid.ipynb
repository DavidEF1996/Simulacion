```python
import numpy as np
import time


rango = int((input("Ingrese la cantidad de numeros pseudoaleatorios:  ")))

a = 74731897457
b = 37747318974
m = 19

cont = 0
lista = []
contInterac = []

for n in range(rango):
    semilla = int(time.time())
    
    print("semilla",semilla,"iteracion",n)
    cont = 0
    while(True):
        
        xn = (a * semilla + b) % m
        semilla = xn 
        div = semilla / m
        
        #print(cont,'\t\t',semilla,'\t',"{0:.4f}".format(div),(semilla in lista) is True)
        
        if((div in lista) is True):
            
            contInterac.append(cont)
            
            break
       
        cont = cont+1
        lista.append(div)
        
        
```

```python
from scipy.stats import chisquare
print(lista)
chisquare(lista, ddof=0.05)
```
